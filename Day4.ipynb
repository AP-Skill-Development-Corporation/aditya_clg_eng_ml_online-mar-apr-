```python
from matplotlib import style
style.use("ggplot") # grid plot
days=[1,2,3,4,5]
East_temp=[36.6,37,37.7,38,40.1]
vij_temp=[39.9,40.2,42.1,39,45]
plt.plot(days,East_temp,'m^--',linewidth=3,markersize=3,label="guntur_temp")
plt.plot(days,vij_temp,'yo:',linewidth=3,markersize=10,label="vij_temp")
plt.title("Tempeartures of Vij and Guntur")
plt.xlabel("days",fontsize=15)
plt.ylabel("temperature",fontsize=15)
plt.legend(loc=1)
plt.show()
```

### Introduction to Machine Learning
- Machine Learning is the study of algorithms which improves the performance of machine without being explicitly programmed
- the machine is learning under the supervision of human 
- 3types of ML
    1. Supervised
        - the machine is learning under the supervision of human
            - both input/featues and ouput/target is provided to the machine
            - evaluation of both original and predicted outcomes
    2. Unsupervised and
        - without the help human/third party by on its own
        - we need to predict the outcome
        - we need to find the accuracy of model what we build
        - video games: ludo/chess/carroms:computer
    3. Reinforcement
        - dynamic environment such as automated cars,text to speech,image recognition,spam detection,traffic detection,medical dignosis etc.
